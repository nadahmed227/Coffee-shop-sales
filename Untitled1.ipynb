```python
import pandas as pd  
import matplotlib.pyplot as plt   # الرسومات البيانية الأساسية
import seaborn as sns             
# قراءة البيانات
df=pd.read_csv('index_1.csv')
```

```python
df['date']=pd.to_datetime(df['date'],errors='coerce')
df['datetime']=pd.to_datetime(df['datetime'] ,format='%I:%M:%S %p',errors='coerce')
df['money']=pd.to_numeric(df['money'],errors='coerce') 
df['month']=df['date'].dt.month 
df['hour']=df['datetime'].dt.month
```

```python
#KPI
total_revenue=df['money'].sum()
print("💰 Total Revenue:", total_revenue) 
total_orders=len(df)
print("📦 Total Orders:", total_orders)
top_coffee=df['coffee_name'].value_counts().head(5)
print("☕ Top 5 Coffees:\n", top_coffee)
Unique_Customers=df['card'].nunique()
print("👥 Unique Customers:", Unique_Customers)
```

```python
#Visualization
# 1️⃣ Payment Methods Distribution
plt.subplot(2,2,1)
df.groupby('cash_type')['money'].sum().plot(kind='bar' ,color=['#4CAF50','#2196F3'])
plt.title("Payment Methods Distribution")
plt.ylabel("Total Sales")
```

```python
# 2️⃣ Sales Trends (Monthly)
plt.subplot(2,2,2)
df.groupby('month')['money'].sum().plot(kind='line',marker='o',color='orange')
plt.title("Sales Trends (Monthly)")
plt.ylabel("Revenue")
```

```python
plt.subplot(2,2,3)
df.groupby('coffee_name')['money'].sum().sort_values(ascending=False).head(5).plot(kind='bar',color='blue')
plt.title("Top 5 Coffee Types")
```

```python
plt.subplot(2,2,4)
df.groupby('hour')['money'].sum().plot(kind='bar',color='purple')
plt.title("Peak Hours")
plt.xlabel("Hour of Day (0-23)")
```

```python
plt.tight_layout()
plt.show()
```
